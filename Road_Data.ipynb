```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\91969\OneDrive\Desktop\Safety_Data.csv")
```

```python
print ("Road   Safety   Data:")
print ( "\nFirst 5 rows of data: ")
display(df.head(5))
```

```python
print ("\nSummary Statics:")

display(df.describe(include = 'all'))
```

```python
plt.figure (figsize =(5,3))
sns.countplot(data = df , x = 'Severity'  , palette ='viridis')
plt.title('Number of Accidents by Severrity')
plt.xlabel('Severity')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize = (4,2))
sns.countplot (data=df , x ='Weather Conditions' , palette = 'coolwarm')
plt.title ('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions ')
plt.ylabel('Number of Accidents')
plt.xticks(rotation =45)
plt.show()
```

```python
plt.figure(figsize =(5,4))
sns.barplot(data=df ,x ='Vehicle Type' , y= 'Number of Casualities')
plt.title ('Total Casualities by Vehicle Type')
plt.xlabel ('Vehicle Type ')
plt.ylabel ( 'Total Casualities ')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (5,4))
sns.countplot(data = df , x = 'Road Type' )
plt.title ('Total Casualities by Road Types ')
plt.xlabel ('Road Type ')
plt.ylabel ('Total Casualities')
plt.show()
```
